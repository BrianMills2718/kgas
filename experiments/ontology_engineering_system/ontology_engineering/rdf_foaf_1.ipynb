```python
# foaf term definitions in rdf format
from rdflib import Graph

# Load FOAF ontology
g = Graph()
g.parse("http://xmlns.com/foaf/0.1/")

# List all terms in FOAF ontology
for s, p, o in g:
    print(f"Subject: {s}, Predicate: {p}, Object: {o}")
```

```python
# ontology
from rdflib import Graph, RDF, RDFS, OWL

# URL of the FOAF ontology
ontology_url = "http://xmlns.com/foaf/0.1/"

# Load the ontology into an RDF graph
g = Graph()
g.parse(ontology_url)


# Function to print classes, properties, and descriptions
def print_ontology(graph):
    print("=== CLASSES ===")
    for s in graph.subjects(RDF.type, OWL.Class):
        print(f"Class: {s}")
        for label in graph.objects(s, RDFS.label):
            print(f"  Label: {label}")
        for comment in graph.objects(s, RDFS.comment):
            print(f"  Comment: {comment}")

    print("\n=== PROPERTIES ===")
    for s in graph.subjects(RDF.type, RDF.Property):
        print(f"Property: {s}")
        for label in graph.objects(s, RDFS.label):
            print(f"  Label: {label}")
        for comment in graph.objects(s, RDFS.comment):
            print(f"  Comment: {comment}")
        for domain in graph.objects(s, RDFS.domain):
            print(f"  Domain: {domain}")
        for range_ in graph.objects(s, RDFS.range):
            print(f"  Range: {range_}")


# Call the function to print FOAF ontology details
print_ontology(g)
```
