```python
# Import the GraphBuilder class
from graphbuilder import GraphBuilder

# Create a new graph builder instance
graph = GraphBuilder()

# Add some parent nodes
graph.add_node(
    "parent1", 
    label="Parent Node 1",
    color="#4287f5",  # Blue
    shape="rectangle",
    properties={"fontSize": 14}
)

graph.add_node(
    "parent2",
    label="Parent Node 2", 
    color="#f54242",  # Red
    shape="rectangle",
    properties={"fontSize": 14}
)

# Add children to parent1
children1 = ["child1_1", "child1_2", "child1_3"]
graph.add_children(
    parent="parent1",
    children=children1,
    color="#87ceeb",  # Sky blue
    properties={"fontSize": 12}
)

# Add children to parent2
children2 = ["child2_1", "child2_2"]
graph.add_children(
    parent="parent2",
    children=children2,
    color="#ff9999",  # Light red
    properties={"fontSize": 12}
)

# Add some edges between nodes
graph.add_edge(
    "child1_1",
    "child2_1",
    label="Connection 1",
    color="#666666",
    thickness=2.0,
    directed=True
)

graph.add_edge(
    "child1_2",
    "child2_2",
    label="Connection 2",
    color="#666666",
    thickness=1.5,
    directed=True
)

# Display the graph
graph.display()
```
