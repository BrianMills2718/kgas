```python
def analyze_network(G):
    # Get strongly connected components
    sccs = list(nx.strongly_connected_components(G))
    print("\nStrongly Connected Components:")
    for i, scc in enumerate(sccs, 1):
        if len(scc) > 1:  # Only show components with multiple nodes
            print(f"Component {i}: {scc}")

    # Calculate simpler centrality measures
    degree_cent = nx.degree_centrality(G)
    between_cent = nx.betweenness_centrality(G)

    # Combine centrality measures
    combined_centrality = {}
    for node in G.nodes():
        combined_centrality[node] = (degree_cent[node] + between_cent[node]) / 2

    print("\nTop 10 Most Central Concepts:")
    for node, cent in sorted(
        combined_centrality.items(), key=lambda x: x[1], reverse=True
    )[:10]:
        print(f"{node}: {cent:.3f}")

    # Find node clusters based on connectivity
    clusters = []
    remaining_nodes = set(G.nodes())

    while remaining_nodes:
        node = next(iter(remaining_nodes))
        cluster = {node} | set(G.predecessors(node)) | set(G.successors(node))
        clusters.append(
            cluster & remaining_nodes
        )  # Only include nodes not yet clustered
        remaining_nodes -= cluster

    print("\nConcept Clusters:")
    for i, cluster in enumerate(clusters, 1):
        print(f"\nCluster {i}:")
        print(list(cluster))

    # Find important paths between key concepts
    top_concepts = sorted(
        combined_centrality.items(), key=lambda x: x[1], reverse=True
    )[:5]
    top_nodes = [node for node, _ in top_concepts]

    print("\nKey Conceptual Paths:")
    for i, source in enumerate(top_nodes):
        for target in top_nodes[i + 1 :]:
            try:
                path = nx.shortest_path(G, source, target)
                print(f"\nPath from {source} to {target}:")
                print(" -> ".join(path))
            except nx.NetworkXNoPath:
                print(f"\nNo path from {source} to {target}")

    return {"centrality": combined_centrality, "clusters": clusters}


# Run the analysis
analysis_results = analyze_network(G)
```

```python
import matplotlib.pyplot as plt
import networkx as nx
import random
import numpy as np


# Assuming we have our clusters from the previous analysis
def visualize_clustered_network(G, clusters):
    # Create color map
    colors = plt.cm.rainbow(np.linspace(0, 1, len(clusters)))
    node_colors = {}

    # Assign colors to nodes based on their cluster
    for i, cluster in enumerate(clusters):
        for node in cluster:
            node_colors[node] = colors[i]

    # Create the plot
    plt.figure(figsize=(24, 18))

    # Use spring layout but increase k for more spacing
    pos = nx.spring_layout(G, k=2, iterations=50)

    # Draw nodes
    for i, cluster in enumerate(clusters):
        nx.draw_networkx_nodes(
            G,
            pos,
            nodelist=list(cluster),
            node_color=[colors[i]] * len(cluster),
            node_size=2000,
        )

    # Draw edges
    nx.draw_networkx_edges(G, pos, edge_color="gray", arrows=True, arrowsize=20)

    # Draw labels
    nx.draw_networkx_labels(G, pos)

    # Draw edge labels
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=8)

    plt.title("Clustered Concept Network")
    plt.axis("off")
    plt.tight_layout()
    plt.show()


# Run visualization

visualize_clustered_network(G, analysis_results["clusters"])
```

```python
def generate_toc(G, centrality_scores, clusters):
    # Sort clusters by highest average centrality
    cluster_importance = []
    for i, cluster in enumerate(clusters):
        avg_centrality = sum(centrality_scores[node] for node in cluster) / len(cluster)
        cluster_importance.append((i, avg_centrality, cluster))

    # Sort clusters by importance
    sorted_clusters = sorted(cluster_importance, key=lambda x: x[1], reverse=True)

    print("Table of Contents:\n")

    # Print main sections based on most central clusters
    for section_num, (cluster_idx, importance, cluster) in enumerate(
        sorted_clusters, 1
    ):
        # Find most central node in cluster to use as section title
        section_head = max(cluster, key=lambda x: centrality_scores[x])

        print(f"{section_num}. {section_head}")

        # For each node in cluster, find its connections to other clusters
        for node in sorted(cluster, key=lambda x: centrality_scores[x], reverse=True):
            if node != section_head:
                print(f"   {section_num}.{section_num}. {node}")

            # Find connections to other clusters
            connections = []
            for neighbor in G.neighbors(node):
                for other_idx, _, other_cluster in sorted_clusters:
                    if neighbor in other_cluster and other_idx != cluster_idx:
                        connections.append(
                            f"{neighbor} ({G.edges[node, neighbor].get('label', '')})"
                        )

            if connections:
                print(f"      Related concepts: {', '.join(connections)}")
        print()


# Generate ToC using our analysis results
generate_toc(G, analysis_results["centrality"], analysis_results["clusters"])
```

```python
def analyze_prerequisite_network(G):
    print("\n=== Network Analysis for ToC Generation ===\n")

    # Check if network is properly directed
    print("Directionality Check:")
    print(f"Network has {len(G.edges())} edges")
    print(f"Network is directed: {nx.is_directed(G)}")
    print(f"Network has cycles: {not nx.is_directed_acyclic_graph(G)}")

    # Find root concepts (those with no prerequisites)
    roots = [n for n in G.nodes() if G.in_degree(n) == 0]
    print("\nRoot Concepts (Starting Points):")
    print(roots)

    # Find leaf concepts (those that don't lead to others)
    leaves = [n for n in G.nodes() if G.out_degree(n) == 0]
    print("\nLeaf Concepts (End Points):")
    print(leaves)

    # Calculate different centrality measures
    in_degree = nx.in_degree_centrality(G)  # concepts that depend on many others
    out_degree = nx.out_degree_centrality(G)  # concepts that many others depend on
    betweenness = nx.betweenness_centrality(G)  # concepts that bridge different areas

    # Identify key concepts based on centrality
    print("\nKey Bridge Concepts (High Betweenness):")
    for node, cent in sorted(betweenness.items(), key=lambda x: x[1], reverse=True)[:5]:
        print(f"{node}: {cent:.3f}")

    # Find main concept pathways
    print("\nMain Concept Pathways:")
    for root in roots:
        for leaf in leaves:
            try:
                paths = list(nx.all_simple_paths(G, root, leaf))
                if paths:
                    print(f"\nFrom {root} to {leaf}:")
                    for path in paths[:2]:  # Show up to 2 paths
                        print(" -> ".join(path))
            except nx.NetworkXNoPath:
                continue

    # Find concept layers (concepts grouped by their distance from roots)
    layers = {}
    for node in G.nodes():
        max_dist = 0
        for root in roots:
            try:
                dist = len(nx.shortest_path(G, root, node)) - 1
                max_dist = max(max_dist, dist)
            except nx.NetworkXNoPath:
                continue
        if max_dist in layers:
            layers[max_dist].append(node)
        else:
            layers[max_dist] = [node]

    print("\nConcept Layers (by prerequisite depth):")
    for depth in sorted(layers.keys()):
        print(f"\nLayer {depth}:")
        print(layers[depth])

    # Generate suggested ToC
    print("\nSuggested Table of Contents:")
    for depth in sorted(layers.keys()):
        print(f"\n{depth}. Layer {depth} Concepts")
        layer_concepts = layers[depth]
        # Sort concepts within layer by out-degree (how many concepts they enable)
        layer_concepts.sort(key=lambda x: out_degree[x], reverse=True)
        for i, concept in enumerate(layer_concepts, 1):
            print(f"  {depth}.{i}. {concept}")
            # Show what this concept enables
            successors = list(G.successors(concept))
            if successors:
                print(f"      Enables understanding of: {', '.join(successors)}")
            # Show what this concept requires
            prerequisites = list(G.predecessors(concept))
            if prerequisites:
                print(f"      Requires understanding of: {', '.join(prerequisites)}")

    return {
        "roots": roots,
        "leaves": leaves,
        "layers": layers,
        "centrality": {
            "in_degree": in_degree,
            "out_degree": out_degree,
            "betweenness": betweenness,
        },
    }


# Run the analysis
analysis_results = analyze_prerequisite_network(G)
```

```python
# intelligent chunker- reads through finds section breaks and then rechunks
from openai import OpenAI
from typing import List, Dict
import re
import asyncio
import nest_asyncio
from tqdm.notebook import tqdm
import json

nest_asyncio.apply()


class AsyncDocumentChunker:
    def __init__(self, api_key: str, initial_chunk_size: int = 2000):
        self.client = OpenAI(api_key=api_key)
        self.initial_chunk_size = initial_chunk_size

    def initial_chunk(self, text: str) -> List[str]:
        chunks = []
        overlap = 200
        words = text.split()
        chunk_size_words = self.initial_chunk_size // 5

        for i in range(0, len(words), chunk_size_words - overlap):
            chunk = " ".join(words[i : i + chunk_size_words])
            chunks.append(chunk)

        return chunks

    async def analyze_chunk(self, chunk: str, i: int) -> Dict:
        prompt = f"""Analyze this text chunk and identify natural section breaks.
Output only valid JSON matching this format:
{{
    "section_breaks": [list of character indices],
    "topics": [list of main topics discussed],
    "structural_position": "beginning" if this seems to start a new section, "middle" if it continues a section, or "end" if it ends a section,
    "incomplete_thoughts": true if the chunk cuts off mid-thought, false otherwise
}}

Text to analyze:
{chunk}"""

        try:
            response = self.client.chat.completions.create(
                model="gpt-4o-mini",
                messages=[
                    {
                        "role": "system",
                        "content": "You are a document structure analyzer that only outputs valid JSON.",
                    },
                    {"role": "user", "content": prompt},
                ],
                temperature=0,
            )

            # Try to parse the response as JSON
            analysis = json.loads(response.choices[0].message.content)
            analysis["chunk_index"] = i
            analysis["text"] = chunk
            print(
                f"Chunk {i}: Position={analysis['structural_position']}, Topics={analysis['topics']}"
            )
            return analysis

        except Exception as e:
            print(f"Error processing chunk {i}: {str(e)}")
            return {
                "chunk_index": i,
                "text": chunk,
                "section_breaks": [],
                "topics": [],
                "structural_position": "middle",
                "incomplete_thoughts": True,
            }

    async def analyze_sections(self, chunks: List[str]) -> List[Dict]:
        tasks = [self.analyze_chunk(chunk, i) for i, chunk in enumerate(chunks)]
        return await asyncio.gather(*tasks)

    def rechunk_by_sections(self, text: str, analyses: List[Dict]) -> List[Dict]:
        sections = []
        current_section = {"text": "", "topics": [], "complete": True}

        # Sort analyses by chunk_index
        analyses.sort(key=lambda x: x["chunk_index"])

        # Force first chunk to be beginning and last to be end
        if analyses:
            analyses[0]["structural_position"] = "beginning"
            analyses[-1]["structural_position"] = "end"

        for analysis in analyses:
            chunk_text = analysis["text"]
            position = analysis["structural_position"]

            # Start new section if:
            # 1. This chunk is marked as beginning
            # 2. OR previous section had an end marker
            if position == "beginning" and current_section["text"]:
                sections.append(current_section)
                current_section = {"text": "", "topics": [], "complete": True}

            # Add content to current section
            current_section["text"] += " " + chunk_text
            if analysis["topics"]:
                current_section["topics"].extend(analysis["topics"])

            if analysis["incomplete_thoughts"]:
                current_section["complete"] = False

            # End section if this chunk is marked as end
            if position == "end":
                sections.append(current_section)
                current_section = {"text": "", "topics": [], "complete": True}

        # Add final section if not empty
        if current_section["text"].strip():
            sections.append(current_section)

        # Clean up sections
        for section in sections:
            section["text"] = re.sub(r"\s+", " ", section["text"]).strip()
            section["topics"] = list(set(section["topics"]))

        return sections


async def main():
    api_key = "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"
    file_path = r"C:\Users\bmills\Downloads\semantic_complexity.txt"

    with open(file_path, "r", encoding="utf-8") as f:
        text = f.read()

    chunker = AsyncDocumentChunker(api_key)
    initial_chunks = chunker.initial_chunk(text)
    print(f"Created {len(initial_chunks)} initial chunks")

    analyses = await chunker.analyze_sections(initial_chunks)
    sections = chunker.rechunk_by_sections(text, analyses)

    print(f"\nFinal number of sections: {len(sections)}")

    # Print results with more detail
    for i, section in enumerate(sections, 1):
        print(f"\nSection {i}:")
        print(f"Topics: {section['topics']}")
        print(f"Complete: {section['complete']}")
        print(f"Text preview: {section['text'][:200]}...")
        print("-" * 80)

    return sections


# Run the async code
sections = await main()
```

```python
from openai import OpenAI
from typing import List, Dict
import re
import asyncio
import nest_asyncio
from tqdm.notebook import tqdm
import json
from collections import Counter

nest_asyncio.apply()


class AsyncDocumentChunker:
    def __init__(self, api_key: str, initial_chunk_size: int = 2000):
        self.client = OpenAI(api_key=api_key)
        self.initial_chunk_size = initial_chunk_size

    def initial_chunk(self, text: str) -> List[str]:
        chunks = []
        overlap = 200
        words = text.split()
        chunk_size_words = self.initial_chunk_size // 5

        for i in range(0, len(words), chunk_size_words - overlap):
            chunk = " ".join(words[i : i + chunk_size_words])
            chunks.append(chunk)

        return chunks

    async def analyze_chunk(self, chunk: str, i: int) -> Dict:
        prompt = f"""Analyze this text chunk and identify the main topic shift. Focus on detecting if this chunk represents a transition to a new major theme or topic.

Output only valid JSON matching this format:
{{
    "main_theme": "the primary theme/topic of this chunk",
    "topics": [list of 3-5 most important subtopics],
    "starts_new_theme": true if this chunk seems to start a new major theme/topic, false otherwise,
    "ends_theme": true if this chunk seems to conclude a theme/topic, false otherwise,
    "incomplete_thoughts": true if the chunk cuts off mid-thought, false otherwise
}}

Text to analyze:
{chunk}"""

        try:
            response = self.client.chat.completions.create(
                model="gpt-4o-mini",
                messages=[
                    {
                        "role": "system",
                        "content": "You are a document structure analyzer that only outputs valid JSON.",
                    },
                    {"role": "user", "content": prompt},
                ],
                temperature=0,
            )

            analysis = json.loads(response.choices[0].message.content)
            analysis["chunk_index"] = i
            analysis["text"] = chunk
            print(
                f"Chunk {i}: Theme={analysis['main_theme']}, New Theme={analysis['starts_new_theme']}"
            )
            return analysis

        except Exception as e:
            print(f"Error processing chunk {i}: {str(e)}")
            return {
                "chunk_index": i,
                "text": chunk,
                "main_theme": "unknown",
                "topics": [],
                "starts_new_theme": False,
                "ends_theme": False,
                "incomplete_thoughts": True,
            }

    async def analyze_sections(self, chunks: List[str]) -> List[Dict]:
        tasks = [self.analyze_chunk(chunk, i) for i, chunk in enumerate(chunks)]
        return await asyncio.gather(*tasks)

    def rechunk_by_sections(self, analyses: List[Dict]) -> List[Dict]:
        sections = []
        current_section = {"text": "", "main_theme": "", "topics": [], "complete": True}

        # Sort analyses by chunk_index
        analyses.sort(key=lambda x: x["chunk_index"])

        # Force first chunk to start new theme
        if analyses:
            analyses[0]["starts_new_theme"] = True

        for analysis in analyses:
            # Start new section if:
            # 1. This chunk starts a new theme
            # 2. OR we've accumulated significant content (>= 3 chunks)
            if (analysis["starts_new_theme"] and current_section["text"]) or (
                len(current_section["text"].split()) >= 300
                and analysis["main_theme"] != current_section["main_theme"]
            ):

                sections.append(current_section)
                current_section = {
                    "text": "",
                    "main_theme": analysis["main_theme"],
                    "topics": [],
                    "complete": True,
                }

            # Add content to current section
            current_section["text"] += " " + analysis["text"]
            current_section["topics"].extend(analysis["topics"])
            current_section["main_theme"] = (
                current_section["main_theme"] or analysis["main_theme"]
            )

            if analysis["incomplete_thoughts"]:
                current_section["complete"] = False

        # Add final section if not empty
        if current_section["text"].strip():
            sections.append(current_section)

        # Clean up sections
        for section in sections:
            section["text"] = re.sub(r"\s+", " ", section["text"]).strip()
            # Keep most common topics
            topic_counts = Counter(section["topics"])
            section["topics"] = [topic for topic, count in topic_counts.most_common(5)]

        return sections


async def main():
    api_key = "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"
    file_path = r"C:\Users\bmills\Downloads\semantic_complexity.txt"

    with open(file_path, "r", encoding="utf-8") as f:
        text = f.read()

    chunker = AsyncDocumentChunker(api_key)
    initial_chunks = chunker.initial_chunk(text)
    print(f"Created {len(initial_chunks)} initial chunks")

    analyses = await chunker.analyze_sections(initial_chunks)
    sections = chunker.rechunk_by_sections(analyses)

    print(f"\nFinal number of sections: {len(sections)}")

    for i, section in enumerate(sections, 1):
        print(f"\nSection {i}:")
        print(f"Main Theme: {section['main_theme']}")
        print(f"Key Topics: {section['topics']}")
        print(f"Complete: {section['complete']}")
        print(f"Text preview: {section['text'][:200]}...")
        print("-" * 80)

    return sections


# Run the async code
sections = await main()
```

```python
import networkx as nx
from pyvis.network import Network
import json
from collections import defaultdict


def create_semantic_network(analyses):
    """Create a network from semantic analyses"""
    G = nx.DiGraph()

    # Track node types for coloring
    node_types = {}

    # Add nodes and edges from each chunk
    for i, analysis in enumerate(analyses):
        # Add core concepts
        for concept in analysis["core_concepts"]:
            if concept not in G:
                G.add_node(concept)
                node_types[concept] = "core"

        # Add supporting concepts
        for concept in analysis["supporting_concepts"]:
            if concept not in G:
                G.add_node(concept)
                node_types[concept] = "supporting"

        # Add relationships as edges
        for rel in analysis["relationships"]:
            G.add_edge(
                rel["from"],
                rel["to"],
                type=rel["type"],
                title=f"{rel['from']} {rel['type']} {rel['to']}",
            )

    # Create interactive visualization
    nt = Network(
        height="750px",
        width="100%",
        bgcolor="#ffffff",
        font_color="black",
        notebook=True,
    )

    # Configure physics
    nt.force_atlas_2based()
    nt.show_buttons(filter_=["physics"])

    # Add nodes with different colors based on type
    for node in G.nodes():
        color = "#ff7676" if node_types.get(node) == "core" else "#76a8ff"
        nt.add_node(
            node,
            label=node,
            color=color,
            title=f"Type: {node_types.get(node, 'unknown')}",
        )

    # Add edges with relationship types
    for edge in G.edges(data=True):
        nt.add_edge(
            edge[0],
            edge[1],
            title=edge[2].get("type", ""),
            label=edge[2].get("type", ""),
        )

    # Generate the HTML file
    nt.save_graph("semantic_network.html")

    return G


def print_network_stats(G):
    """Print basic statistics about the network"""
    print("\nNetwork Statistics:")
    print(f"Number of nodes: {G.number_of_nodes()}")
    print(f"Number of edges: {G.number_of_edges()}")

    # Most central concepts
    centrality = nx.betweenness_centrality(G)
    print("\nMost central concepts:")
    for node, score in sorted(centrality.items(), key=lambda x: x[1], reverse=True)[:5]:
        print(f"- {node}: {score:.3f}")

    # Most connected concepts
    degrees = dict(G.degree())
    print("\nMost connected concepts:")
    for node, degree in sorted(degrees.items(), key=lambda x: x[1], reverse=True)[:5]:
        print(f"- {node}: {degree} connections")

    # Most common relationship types
    relationship_counts = defaultdict(int)
    for _, _, data in G.edges(data=True):
        if "type" in data:
            relationship_counts[data["type"]] += 1

    print("\nMost common relationship types:")
    for rel_type, count in sorted(
        relationship_counts.items(), key=lambda x: x[1], reverse=True
    ):
        print(f"- {rel_type}: {count}")


# Create and analyze the network
G = create_semantic_network(analyses)
print_network_stats(G)

print("\nVisualization saved as 'semantic_network.html'")
```

```python
import networkx as nx
from pyvis.network import Network
import json
from collections import defaultdict
from openai import OpenAI
import asyncio
import nest_asyncio
import re

nest_asyncio.apply()


async def get_natural_breaks(text: str, client: OpenAI) -> str:
    prompt = """Read this conversation and identify if/where there are any clear, natural breaking points where one major topic or theme concludes and another begins. 

Only identify breaks if they are very clear and natural - if in doubt, prefer to keep the text as one continuous piece.

Output your response in this format:
1. Are there any clear natural breaks? (yes/no)
2. If yes, paste the text immediately before and after each break point to show the transition"""

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {
                "role": "system",
                "content": "You are helping to analyze the natural flow of a conversation.",
            },
            {"role": "user", "content": f"{prompt}\n\nText to analyze:\n{text}"},
        ],
        temperature=0,
    )

    return response.choices[0].message.content


async def get_chunks_from_breaks(breaks_text: str) -> list:
    chunks = []
    break_pattern = r"- Before: \"(.*?)\"[\s\n]*- After: \"(.*?)\""
    matches = re.finditer(break_pattern, breaks_text, re.DOTALL)

    for match in matches:
        before = match.group(1).strip()
        after = match.group(2).strip()
        chunks.append(f"{before} {after}")

    return chunks


async def analyze_chunk_semantics(chunk: str, client: OpenAI) -> dict:
    prompt = f"""Analyze this conversation chunk and identify its semantic structure.
Text to analyze: {chunk}

Provide your analysis in this exact JSON format:
{{
    "core_concepts": ["list of main topics/concepts"],
    "supporting_concepts": ["list of subtopics/supporting ideas"],
    "examples": ["list of examples or evidence"],
    "relationships": [
        {{
            "from": "concept1",
            "to": "concept2",
            "type": "one of: is_subtopic_of, exemplifies, defines, explains, supports, contradicts, implies"
        }}
    ]
}}"""

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {
                "role": "system",
                "content": "You are a semantic structure analyzer. Output only valid JSON.",
            },
            {"role": "user", "content": prompt},
        ],
        temperature=0,
    )

    return json.loads(response.choices[0].message.content)


def create_semantic_network(analyses, chunks):
    G = nx.DiGraph()

    # Initialize node tracking
    node_data = {}

    # Process each chunk's analysis
    for i, (analysis, chunk_text) in enumerate(zip(analyses, chunks)):
        # Add core concepts
        for concept in analysis["core_concepts"]:
            if concept not in node_data:
                node_data[concept] = {"type": "core", "chunks": [], "chunk_texts": []}
            node_data[concept]["chunks"].append(i)
            node_data[concept]["chunk_texts"].append(chunk_text)

        # Add supporting concepts
        for concept in analysis["supporting_concepts"]:
            if concept not in node_data:
                node_data[concept] = {
                    "type": "supporting",
                    "chunks": [],
                    "chunk_texts": [],
                }
            node_data[concept]["chunks"].append(i)
            node_data[concept]["chunk_texts"].append(chunk_text)

        # Add relationships
        for rel in analysis["relationships"]:
            # Ensure both nodes exist
            for node in [rel["from"], rel["to"]]:
                if node not in node_data:
                    node_data[node] = {
                        "type": "supporting",
                        "chunks": [i],
                        "chunk_texts": [chunk_text],
                    }

            # Add the edge
            G.add_edge(
                rel["from"],
                rel["to"],
                type=rel["type"],
                chunk_index=i,
                chunk_text=chunk_text,
            )

    # Add all nodes with their data
    for node, data in node_data.items():
        G.add_node(node, **data)

    return G


def visualize_network(G):
    nt = Network(
        height="750px",
        width="100%",
        bgcolor="#ffffff",
        font_color="black",
        notebook=True,
        cdn_resources="remote",
    )

    nt.force_atlas_2based()
    nt.show_buttons(filter_=["physics"])

    # Add nodes
    for node, data in G.nodes(data=True):
        # Get node type with fallback
        node_type = data.get("type", "unknown")
        color = "#ff7676" if node_type == "core" else "#76a8ff"

        # Create hover text
        chunks = data.get("chunks", [])
        title = f"Concept: {node}\nType: {node_type}\n"
        title += f"Appears in {len(chunks)} chunks"

        nt.add_node(node, label=node, color=color, title=title)

    # Add edges
    for edge in G.edges(data=True):
        title = f"{edge[0]} {edge[2].get('type', 'relates to')} {edge[1]}\n"
        title += f"From chunk {edge[2].get('chunk_index', '?') + 1}"

        nt.add_edge(
            edge[0], edge[1], title=title, label=edge[2].get("type", "relates to")
        )

    nt.save_graph("semantic_network.html")


def analyze_network_for_essay(G):
    print("\nNetwork Analysis for Essay Generation:")

    # Convert to undirected for community detection
    G_undir = G.to_undirected()
    communities = nx.community.louvain_communities(G_undir)
    print(f"\nIdentified {len(communities)} main topic clusters:")

    # Calculate betweenness centrality once
    centrality = nx.betweenness_centrality(G)

    clusters = []
    for i, community in enumerate(communities):
        relevant_chunks = set()
        community_concepts = list(community)

        # Get all relevant chunks for this community
        for concept in community:
            node_data = G.nodes[concept]
            if "chunks" in node_data:
                relevant_chunks.update(node_data["chunks"])

        # Find most central concept in this community
        community_centrality = {node: centrality[node] for node in community}
        central_concept = max(community_centrality.items(), key=lambda x: x[1])[0]

        cluster = {
            "id": i,
            "concepts": community_concepts,
            "chunk_indices": sorted(list(relevant_chunks)),
            "central_concept": central_concept,
        }
        clusters.append(cluster)

        print(f"\nCluster {i + 1} - Central concept: {central_concept}")
        print(f"Key concepts: {', '.join(community_concepts[:5])}")
        print(f"Spans chunks: {cluster['chunk_indices']}")

    return clusters


def get_cluster_context(G, cluster):
    context = {
        "central_concept": cluster["central_concept"],
        "concepts": cluster["concepts"],
        "chunks": [],
        "relationships": [],
    }

    for concept in cluster["concepts"]:
        node_data = G.nodes[concept]
        for chunk_text in node_data["chunk_texts"]:
            if chunk_text not in context["chunks"]:
                context["chunks"].append(chunk_text)

        for _, target, data in G.edges(concept, data=True):
            if target in cluster["concepts"]:
                context["relationships"].append(
                    {"from": concept, "to": target, "type": data["type"]}
                )

    return context


async def main():
    api_key = "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"
    file_path = r"C:\Users\bmills\Downloads\semantic_complexity.txt"

    client = OpenAI(api_key=api_key)

    with open(file_path, "r", encoding="utf-8") as f:
        text = f.read()

    # Get breaks and chunks
    breaks_analysis = await get_natural_breaks(text, client)
    chunks = await get_chunks_from_breaks(breaks_analysis)
    print(f"Found {len(chunks)} chunks")

    # Analyze chunks
    analyses = []
    for i, chunk in enumerate(chunks, 1):
        print(f"Analyzing chunk {i}...")
        analysis = await analyze_chunk_semantics(chunk, client)
        analyses.append(analysis)

    # Create and analyze network
    G = create_semantic_network(analyses, chunks)
    visualize_network(G)

    # Get clusters and analyze them
    clusters = analyze_network_for_essay(G)

    if clusters:
        print("\nExample context for first cluster:")
        context = get_cluster_context(G, clusters[0])
        print(f"Central concept: {context['central_concept']}")
        print(f"Number of relevant chunks: {len(context['chunks'])}")
        print(f"Number of internal relationships: {len(context['relationships'])}")

    return G, clusters, chunks, analyses


# Run the pipeline
G, clusters, chunks, analyses = await main()
```

```python
import matplotlib.pyplot as plt
import networkx as nx
import numpy as np


def visualize_network_matplotlib(G):
    # Create figure
    plt.figure(figsize=(20, 20))

    # Set up layout
    pos = nx.spring_layout(G, k=1, iterations=50)

    # Draw nodes
    core_nodes = [
        node for node, attr in G.nodes(data=True) if attr.get("type") == "core"
    ]
    support_nodes = [
        node for node, attr in G.nodes(data=True) if attr.get("type") == "supporting"
    ]

    # Draw edges with different colors based on relationship type
    edge_colors = []
    edge_labels = {}
    for u, v, data in G.edges(data=True):
        rel_type = data.get("type", "unknown")
        edge_labels[(u, v)] = rel_type

    # Draw the network
    nx.draw_networkx_nodes(
        G,
        pos,
        nodelist=core_nodes,
        node_color="red",
        node_size=2000,
        alpha=0.6,
        label="Core Concepts",
    )
    nx.draw_networkx_nodes(
        G,
        pos,
        nodelist=support_nodes,
        node_color="blue",
        node_size=1500,
        alpha=0.6,
        label="Supporting Concepts",
    )

    # Draw edges
    nx.draw_networkx_edges(G, pos, alpha=0.5, arrows=True, edge_color="gray", width=1)

    # Add labels
    nx.draw_networkx_labels(G, pos, font_size=8)

    # Add edge labels
    nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=6)

    plt.title("Semantic Network Analysis", fontsize=16)
    plt.legend(fontsize=12)
    plt.axis("off")

    # Save and show
    plt.savefig("semantic_network.png", dpi=300, bbox_inches="tight")
    plt.show()


def visualize_clusters(G, clusters):
    """Visualize the clusters separately"""
    # Create a figure with subplots for each major cluster
    major_clusters = [
        c for c in clusters if len(c["concepts"]) > 3
    ]  # Only show larger clusters
    n_clusters = len(major_clusters)

    if n_clusters > 0:
        rows = int(np.ceil(n_clusters / 2))
        fig, axes = plt.subplots(rows, 2, figsize=(20, 10 * rows))
        axes = axes.flatten() if rows > 1 else [axes]

        for i, cluster in enumerate(major_clusters):
            if i < len(axes):
                # Create subgraph for this cluster
                subgraph = G.subgraph(cluster["concepts"])
                pos = nx.spring_layout(subgraph)

                # Draw on the appropriate subplot
                ax = axes[i]
                nx.draw(
                    subgraph,
                    pos,
                    ax=ax,
                    with_labels=True,
                    node_color="lightblue",
                    node_size=1000,
                    font_size=8,
                    arrows=True,
                )
                ax.set_title(f"Cluster {i+1}: {cluster['central_concept']}")

        plt.tight_layout()
        plt.savefig("semantic_clusters.png", dpi=300, bbox_inches="tight")
        plt.show()


# Visualize the full network
visualize_network_matplotlib(G)

# Visualize individual clusters
visualize_clusters(G, clusters)
```

```python
def analyze_for_toc(G, clusters):
    """Analyze network structure to generate hierarchical ToC using simple metrics"""
    print("Calculating network metrics...")

    # Calculate simpler centrality measures
    betweenness = nx.betweenness_centrality(G)
    degree_centrality = nx.degree_centrality(G)
    in_degree = dict(G.in_degree())
    out_degree = dict(G.out_degree())

    # Structure for ToC
    toc_structure = {
        "main_topics": [],
        "subtopics": defaultdict(list),
        "support_material": defaultdict(list),
    }

    print("\nAnalyzing topic hierarchy...")

    # Find main topics (using betweenness and degree)
    for node in G.nodes():
        score = betweenness.get(node, 0) * 0.5 + degree_centrality.get(node, 0) * 0.5

        # Check if this node is a central concept in any cluster
        is_central = any(cluster.get("central_concept") == node for cluster in clusters)

        if score > 0.1 or is_central:  # Threshold for main topics
            toc_structure["main_topics"].append(
                {
                    "topic": node,
                    "score": score,
                    "in_degree": in_degree.get(node, 0),
                    "out_degree": out_degree.get(node, 0),
                    "betweenness": betweenness.get(node, 0),
                    "is_central": is_central,
                }
            )

    # Sort main topics by score
    toc_structure["main_topics"].sort(key=lambda x: x["score"], reverse=True)

    print("\nAnalyzing subtopic relationships...")

    # Find subtopics for each main topic
    for main_topic in toc_structure["main_topics"]:
        topic = main_topic["topic"]

        # Get direct successors (nodes that this topic points to)
        successors = list(G.successors(topic))

        # Filter and score subtopics
        for subtopic in successors:
            edge_data = G.get_edge_data(topic, subtopic)
            relationship = edge_data.get("type", "")

            if relationship in ["is_subtopic_of", "explains", "defines", "supports"]:
                subtopic_score = (
                    betweenness.get(subtopic, 0) * 0.5
                    + degree_centrality.get(subtopic, 0) * 0.5
                )

                toc_structure["subtopics"][topic].append(
                    {
                        "subtopic": subtopic,
                        "score": subtopic_score,
                        "relationship": relationship,
                    }
                )

    print("\nGenerating Table of Contents...")

    # Generate ToC text
    toc_text = "# Table of Contents\n\n"

    for i, main_topic in enumerate(toc_structure["main_topics"], 1):
        topic = main_topic["topic"]
        toc_text += f"{i}. {topic}\n"

        # Add subtopics
        for j, subtopic_data in enumerate(
            sorted(
                toc_structure["subtopics"][topic],
                key=lambda x: x["score"],
                reverse=True,
            ),
            1,
        ):
            subtopic = subtopic_data["subtopic"]
            relationship = subtopic_data["relationship"]
            toc_text += f"   {i}.{j}. {subtopic} ({relationship})\n"

            # Get supporting concepts for this subtopic
            supporters = [
                (v, d["type"])
                for u, v, d in G.edges(subtopic, data=True)
                if d["type"] in ["supports", "exemplifies", "illustrates"]
            ]

            for k, (supporter, rel_type) in enumerate(supporters, 1):
                toc_text += f"      {i}.{j}.{k}. {supporter} ({rel_type})\n"

    # Print analysis details
    print("\nNetwork Statistics:")
    print(f"Number of nodes: {G.number_of_nodes()}")
    print(f"Number of edges: {G.number_of_edges()}")
    print(f"Number of main topics: {len(toc_structure['main_topics'])}")
    print(
        f"Average subtopics per main topic: {sum(len(v) for v in toc_structure['subtopics'].values())/len(toc_structure['main_topics']) if toc_structure['main_topics'] else 0:.2f}"
    )

    print("\nProposed Table of Contents:")
    print(toc_text)

    return toc_structure, toc_text


# Get ToC
toc_structure, toc_text = analyze_for_toc(G, clusters)
```

```python
from openai import OpenAI
import asyncio
from typing import Dict, List
import networkx as nx


class EssayGenerator:
    def __init__(
        self, api_key: str, G: nx.DiGraph, toc_structure: Dict, chunks: List[str]
    ):
        self.client = OpenAI(api_key=api_key)
        self.G = G
        self.toc_structure = toc_structure
        self.chunks = chunks

    def get_context_for_topic(self, topic: str) -> Dict:
        """Gather all relevant context for a topic"""
        context = {
            "topic": topic,
            "relevant_chunks": [],
            "related_concepts": [],
            "relationships": [],
            "supporting_evidence": [],
        }

        # Get chunks where this topic appears
        node_data = self.G.nodes[topic]
        if "chunk_texts" in node_data:
            context["relevant_chunks"].extend(node_data["chunk_texts"])

        # Get directly connected concepts and their relationships
        for _, neighbor, data in self.G.edges(topic, data=True):
            context["relationships"].append(
                {"from": topic, "to": neighbor, "type": data.get("type", "related to")}
            )
            context["related_concepts"].append(neighbor)

            # Get chunks for connected concepts too
            neighbor_data = self.G.nodes[neighbor]
            if "chunk_texts" in neighbor_data:
                context["relevant_chunks"].extend(neighbor_data["chunk_texts"])

        # Get supporting evidence (incoming edges)
        for source, _, data in self.G.in_edges(topic, data=True):
            if data.get("type") in ["supports", "exemplifies", "illustrates"]:
                context["supporting_evidence"].append(
                    {"concept": source, "type": data.get("type")}
                )

        return context

    async def generate_chapter(self, main_topic: Dict) -> str:
        """Generate a chapter for a main topic"""
        topic = main_topic["topic"]
        context = self.get_context_for_topic(topic)

        # Get subtopics and their context
        subtopics = self.toc_structure["subtopics"].get(topic, [])
        for subtopic in subtopics:
            subtopic_context = self.get_context_for_topic(subtopic["subtopic"])
            context["relevant_chunks"].extend(subtopic_context["relevant_chunks"])

        # Create detailed prompt for chapter generation
        prompt = f"""Write a detailed essay chapter about '{topic}'. 

Context:
- Main topic: {topic}
- Related concepts: {', '.join(context['related_concepts'])}
- Key relationships: {context['relationships']}
- Supporting evidence: {context['supporting_evidence']}

Relevant text from discussion:
{' '.join(set(context['relevant_chunks']))}

Subtopics to cover:
{', '.join(s['subtopic'] for s in subtopics)}

Write a coherent chapter that:
1. Introduces the main topic
2. Develops each subtopic
3. Uses the provided evidence and relationships
4. Maintains logical flow between sections
5. Integrates the original discussion points

Format using markdown with proper section headers."""

        response = self.client.chat.completions.create(
            model="gpt-4o-mini",
            messages=[
                {
                    "role": "system",
                    "content": "You are writing an academic essay that synthesizes complex discussions into clear, well-structured chapters.",
                },
                {"role": "user", "content": prompt},
            ],
            temperature=0.7,
            max_tokens=2000,
        )

        return response.choices[0].message.content

    async def generate_essay(self) -> str:
        """Generate the complete essay"""
        print("Generating essay chapters...")

        essay_parts = ["# Complete Essay\n\n"]

        for i, main_topic in enumerate(self.toc_structure["main_topics"], 1):
            print(f"\nGenerating Chapter {i}: {main_topic['topic']}...")
            chapter_text = await self.generate_chapter(main_topic)
            essay_parts.append(f"\n## Chapter {i}: {main_topic['topic']}\n\n")
            essay_parts.append(chapter_text)
            essay_parts.append("\n---\n")

        complete_essay = "\n".join(essay_parts)

        # Save essay to file
        with open("generated_essay.md", "w", encoding="utf-8") as f:
            f.write(complete_essay)

        print("\nEssay generation complete! Saved to 'generated_essay.md'")
        return complete_essay


async def main():
    api_key = "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"

    generator = EssayGenerator(api_key, G, toc_structure, chunks)
    essay = await generator.generate_essay()

    # Print first 500 characters as preview
    print("\nEssay Preview:")
    print(essay[:500] + "...")


# Generate the essay
await main()
```

```python
from openai import OpenAI
import asyncio


async def generate_chapter_summary(client: OpenAI, topic: str, context: dict) -> str:
    """Generate a simple summary of a chapter's content at 10th grade level"""

    prompt = f"""Summarize the main ideas about '{topic}' in simple terms that a 10th grade student would understand.

Key points to cover:
- What is {topic}?
- How does it connect to {', '.join(context['related_concepts'][:3])}?
- What are the most important ideas?

Use these text chunks for context:
{' '.join(set(context['relevant_chunks']))}

Rules:
1. Write 3-4 paragraphs
2. Use simple language
3. Give concrete examples where possible
4. Avoid technical jargon unless you explain it
5. Break complex ideas into simpler parts"""

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {
                "role": "system",
                "content": "You are writing chapter summaries for high school students.",
            },
            {"role": "user", "content": prompt},
        ],
        temperature=0.7,
        max_tokens=750,
    )

    return response.choices[0].message.content


async def summarize_all_chapters(G, toc_structure, api_key: str):
    """Generate summaries for all chapters"""
    client = OpenAI(api_key=api_key)

    print("Generating chapter summaries...\n")
    summaries = []

    for i, main_topic in enumerate(toc_structure["main_topics"], 1):
        topic = main_topic["topic"]
        print(f"Summarizing Chapter {i}: {topic}...")

        # Get context from graph
        context = {"relevant_chunks": [], "related_concepts": []}

        # Get chunks where this topic appears
        if "chunk_texts" in G.nodes[topic]:
            context["relevant_chunks"] = G.nodes[topic]["chunk_texts"]

        # Get directly connected concepts
        context["related_concepts"] = [n for n in G.neighbors(topic)]

        # Generate summary
        summary = await generate_chapter_summary(client, topic, context)

        summaries.append(f"\nChapter {i}: {topic}\n{'='*40}\n{summary}\n")

    # Save summaries
    full_text = (
        "\nChapter Summaries (10th Grade Level)\n"
        + "=" * 50
        + "\n"
        + "\n".join(summaries)
    )

    with open("chapter_summaries.md", "w", encoding="utf-8") as f:
        f.write(full_text)

    print("\nSummaries saved to 'chapter_summaries.md'")
    return full_text


# Run summarizer
api_key = "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"
summaries = await summarize_all_chapters(G, toc_structure, api_key)

# Print preview
print("\nPreview of first summary:")
print(summaries.split("\nChapter")[1])
```

```python
import networkx as nx
from collections import defaultdict


def create_consolidated_toc(G, clusters, max_chapters=6):
    """Create a consolidated ToC with fewer main topics"""

    # Calculate node importance
    betweenness = nx.betweenness_centrality(G)
    degree_centrality = nx.degree_centrality(G)

    # Group related clusters by shared concepts and edges
    consolidated_topics = []
    used_cluster_ids = set()  # Track clusters by their index instead

    # First, identify the most central concepts across clusters
    cluster_scores = []
    for i, cluster in enumerate(clusters):
        if len(cluster["concepts"]) < 2:  # Skip tiny clusters
            continue

        # Calculate cluster importance
        cluster_score = sum(
            betweenness.get(concept, 0) * 0.5 + degree_centrality.get(concept, 0) * 0.5
            for concept in cluster["concepts"]
        )

        cluster_scores.append(
            {
                "cluster": cluster,
                "score": cluster_score,
                "central_concept": cluster["central_concept"],
                "cluster_id": i,  # Add cluster ID
            }
        )

    # Sort clusters by importance
    cluster_scores.sort(key=lambda x: x["score"], reverse=True)

    # Take top N clusters as seed topics
    seed_clusters = cluster_scores[:max_chapters]

    # For each seed cluster, absorb related smaller clusters
    for seed in seed_clusters:
        topic = {
            "main_concept": seed["central_concept"],
            "concepts": set(seed["cluster"]["concepts"]),
            "subclusters": [seed["cluster"]],
            "score": seed["score"],
        }
        used_cluster_ids.add(seed["cluster_id"])

        # Look for related clusters to absorb
        for other in cluster_scores:
            if (
                other["cluster_id"] == seed["cluster_id"]
                or other["cluster_id"] in used_cluster_ids
            ):
                continue

            # Check for concept overlap or strong connections
            overlap = len(set(other["cluster"]["concepts"]) & topic["concepts"])
            connections = sum(
                1
                for c1 in topic["concepts"]
                for c2 in other["cluster"]["concepts"]
                if G.has_edge(c1, c2) or G.has_edge(c2, c1)
            )

            if overlap > 0 or connections > 1:
                topic["concepts"].update(other["cluster"]["concepts"])
                topic["subclusters"].append(other["cluster"])
                used_cluster_ids.add(other["cluster_id"])

        consolidated_topics.append(topic)

    # Create new ToC structure
    toc_structure = {
        "main_topics": [],
        "subtopics": defaultdict(list),
        "support_material": defaultdict(list),
    }

    # Convert consolidated topics to ToC structure
    for topic in consolidated_topics:
        main_topic = {"topic": topic["main_concept"], "score": topic["score"]}
        toc_structure["main_topics"].append(main_topic)

        # Add subtopics based on cluster concepts and relationships
        for concept in topic["concepts"]:
            if concept != topic["main_concept"]:
                # Determine relationship type from graph
                edge_data = (
                    G.get_edge_data(topic["main_concept"], concept)
                    or G.get_edge_data(concept, topic["main_concept"])
                    or {"type": "related_to"}
                )

                toc_structure["subtopics"][topic["main_concept"]].append(
                    {
                        "subtopic": concept,
                        "relationship": edge_data.get("type", "related_to"),
                        "score": betweenness.get(concept, 0) * 0.5
                        + degree_centrality.get(concept, 0) * 0.5,
                    }
                )

    # Print new ToC
    print("\nConsolidated Table of Contents:")
    for i, main_topic in enumerate(toc_structure["main_topics"], 1):
        topic = main_topic["topic"]
        print(f"\n{i}. {topic}")

        # Sort subtopics by score
        subtopics = sorted(
            toc_structure["subtopics"][topic], key=lambda x: x["score"], reverse=True
        )

        # Print top subtopics
        for j, subtopic in enumerate(
            subtopics[:5], 1
        ):  # Show top 5 subtopics per chapter
            print(f"   {i}.{j}. {subtopic['subtopic']} ({subtopic['relationship']})")

    return toc_structure


# Generate consolidated ToC
consolidated_toc = create_consolidated_toc(G, clusters, max_chapters=6)
```

```python
from openai import OpenAI
import asyncio
from collections import defaultdict


class ConsolidatedEssayGenerator:
    def __init__(
        self, api_key: str, G: nx.DiGraph, consolidated_toc: dict, original_chunks: list
    ):
        self.client = OpenAI(api_key=api_key)
        self.G = G
        self.toc = consolidated_toc
        self.chunks = original_chunks

    def get_context_for_topic(self, main_topic: str, subtopics: list) -> dict:
        """Gather all relevant context for a topic and its subtopics"""
        context = {
            "main_topic": main_topic,
            "relevant_chunks": set(),  # Using set to avoid duplicates
            "concept_relationships": [],
            "subtopic_info": defaultdict(list),
        }

        # Get all relevant nodes
        nodes_to_check = [main_topic] + [s["subtopic"] for s in subtopics]

        # Gather chunks and relationships for each node
        for node in nodes_to_check:
            # Get node data including chunk texts
            node_data = self.G.nodes[node]
            if "chunk_texts" in node_data:
                context["relevant_chunks"].update(node_data["chunk_texts"])

            # Get incoming and outgoing relationships
            for _, target, data in self.G.edges(node, data=True):
                if target in nodes_to_check:
                    context["concept_relationships"].append(
                        {
                            "from": node,
                            "to": target,
                            "type": data.get("type", "related_to"),
                        }
                    )

            for source, _, data in self.G.in_edges(node, data=True):
                if source in nodes_to_check:
                    context["concept_relationships"].append(
                        {
                            "from": source,
                            "to": node,
                            "type": data.get("type", "related_to"),
                        }
                    )

        return context

    async def generate_chapter(self, main_topic: dict, subtopics: list) -> str:
        """Generate a single chapter"""
        topic = main_topic["topic"]
        context = self.get_context_for_topic(topic, subtopics)

        # Create detailed prompt for chapter generation
        prompt = f"""Write a detailed essay chapter about '{topic}'.

Chapter Structure:
1. {topic} (main topic)
{chr(10).join(f'   - {s["subtopic"]} ({s["relationship"]})' for s in subtopics[:5])}

Context from original discussion:
{chr(10).join(sorted(context['relevant_chunks']))}

Key relationships between concepts:
{chr(10).join(f"- {r['from']} {r['type']} {r['to']}" for r in context['concept_relationships'])}

Write a coherent chapter that:
1. Introduces {topic} and its significance
2. Develops each subtopic, showing how they relate to the main topic
3. Uses the original discussion context to maintain accurate representation
4. Integrates the relationships between concepts
5. Provides a clear flow between sections
6. Maintains academic tone while being accessible

Format using markdown with clear section headers."""

        response = self.client.chat.completions.create(
            model="gpt-4o-mini",
            messages=[
                {
                    "role": "system",
                    "content": "You are writing an academic essay that synthesizes complex discussions into clear, well-structured chapters.",
                },
                {"role": "user", "content": prompt},
            ],
            temperature=0.7,
            max_tokens=3000,
        )

        return response.choices[0].message.content

    async def generate_complete_essay(self) -> str:
        """Generate the complete essay with all chapters"""
        print("Generating essay from consolidated structure...")

        essay_parts = [
            "# Semantic Complexity and Perception: A Network Analysis\n\n",
            "## Abstract\n\n",
            "This essay explores the relationship between semantic complexity, perception, and cognitive structures through a network analysis of key concepts and their relationships.\n\n",
        ]

        # Generate each chapter
        for i, main_topic in enumerate(self.toc["main_topics"], 1):
            print(f"\nGenerating Chapter {i}: {main_topic['topic']}...")

            subtopics = self.toc["subtopics"][main_topic["topic"]]
            chapter_text = await self.generate_chapter(main_topic, subtopics)

            essay_parts.append(f"\n## Chapter {i}: {main_topic['topic']}\n\n")
            essay_parts.append(chapter_text)
            essay_parts.append("\n---\n")

        complete_essay = "\n".join(essay_parts)

        # Save essay
        with open("consolidated_essay.md", "w", encoding="utf-8") as f:
            f.write(complete_essay)

        print("\nEssay generation complete! Saved to 'consolidated_essay.md'")
        return complete_essay


async def main():
    api_key = "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"

    generator = ConsolidatedEssayGenerator(api_key, G, consolidated_toc, chunks)
    essay = await generator.generate_complete_essay()

    # Print preview
    print("\nFirst 500 characters of generated essay:")
    print(essay[:500] + "...")


# Generate the essay
await main()
```
